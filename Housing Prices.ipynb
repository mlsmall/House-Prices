## Data Preprocessing

Before we can train our model with a scikit learn algorithm, we have to fill in empty values in our dataset and convert our categorical data (text) into numerical data.

The input variables (Xtrain) include all the columns in this dataset except for the "SalePrice" column.  Sales Price is our target or ytrain variable.

```python
Xtrain = train.drop('SalePrice', axis=1)
Xtest = test
```

```python
ytrain = train['SalePrice']
```

```python
# First we combine both matrices into one dataset
Xtrain['combined column'] = 1
Xtest['combined column'] = 0

dataset = pd.concat([Xtrain, Xtest])
dataset.head()
```

```python
# The interpolate function fills in empty values in numerical columns with an interpolation with respect to other values
# in that column.
dataset=dataset.interpolate()
```

```python
dataset.head(10)
```

```python
# Now we use the function pd.get_dummies to change categorical text variables into numerical values (1 or 0).
dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
dataset.info()
```

Now we split the dataset into train and test sets again

```python
Xtrain = dataset[dataset['combined column']==1]
Xtrain = Xtrain.drop('combined column', axis=1)

Xtest = dataset[dataset['combined column']==0]
Xtest = Xtest.drop('combined column', axis=1)
```

## Modeling

Now on to the model. We will use Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators = 200) # n_estimators is the number of trees in the forest
```

```python
# We fit the model to our training data.
model.fit(Xtrain,ytrain)
```

```python
# now we create our column of predictors using the test data
predictors = model.predict(Xtest)
```

```python
# We make a file to upload the predictors into the Kaggle competition
submission = pd.DataFrame({'Id': test.index, 'SalePrice': predictors})
submission.to_csv('submission.csv', index=False)
```
